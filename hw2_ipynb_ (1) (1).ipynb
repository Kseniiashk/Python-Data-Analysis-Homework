9. Создайте столбец с датой контакта. Дата должна соответствовать информации из столбцов month и day_of_week и относиться к 2021 году (в этом задании может быть, разумеется, много верных ответов, выведите любой)

```python
from datetime import datetime, timedelta
import calendar

month_dict = {
    'jan': 'January', 'feb': 'February', 'mar': 'March', 'apr': 'April', 'may': 'May', 'jun': 'June', 'jul': 'July', 'aug': 'August', 'sep': 'September', 'oct': 'October', 'nov': 'November', 'dec': 'December'
}
day_dict = {
    'mon': 0, 'tue': 1, 'wed': 2, 'thu': 3, 'fri': 4, 'sat': 5, 'sun': 6
}

year = 2021

def contact_date(row):
    firstDay = datetime(year, list(calendar.month_name).index(month_dict[row['month']]), 1)
    return firstDay + timedelta(days=(day_dict[row['day_of_week']] - firstDay.weekday() + 7))

df['contact_date'] = df.apply(contact_date, axis=1)
df
```
